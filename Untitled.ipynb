# Melanoma Detection Project

This notebook uses the `MelanomaPredictor` module to classify skin lesions as **Benign** or **Malignant**.


```python
import os
import matplotlib.pyplot as plt
from PIL import Image
from melanoma_predictor import MelanomaPredictor

# Initialize the predictor (Loads the model)
predictor = MelanomaPredictor('skin.h5')
```

```python
def show_prediction(img_path):
    """Predicts and displays the image with its label."""
    result = predictor.predict(img_path)
    
    if 'error' in result:
        print(f"Error: {result['error']}")
        return
        
    # Load image for display
    img = Image.open(img_path)
    
    plt.figure(figsize=(5, 5))
    plt.imshow(img)
    plt.axis('off')
    
    # Color code: Red for Malignant, Green for Benign
    color = 'red' if result['label'] == 'MALIGNANT' else 'green'
    plt.title(f"{result['label']}\nConfidence: {result['confidence']:.2%}", color=color, fontsize=14)
    plt.show()
```

```python
# Run Demo on random Test Images
import random

test_dir = 'dataset/test'
if os.path.exists(test_dir):
    benign_files = [os.path.join(test_dir, 'benign', f) for f in os.listdir(os.path.join(test_dir, 'benign')) if f.endswith('.jpg')]
    malignant_files = [os.path.join(test_dir, 'malignant', f) for f in os.listdir(os.path.join(test_dir, 'malignant')) if f.endswith('.jpg')]
    
    print("--- Predicting Random Benign Sample ---")
    if benign_files: show_prediction(random.choice(benign_files))
    
    print("\n--- Predicting Random Malignant Sample ---")
    if malignant_files: show_prediction(random.choice(malignant_files))
```
